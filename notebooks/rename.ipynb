```python
def changeDateFormat():
    client_id = Utils.fromConfig('mongo_client')
    data_base_name = Utils.fromConfig('data_base_name')
    DB = Database(
    client_id=client_id, data_base_name=data_base_name, enforce_coll_name=False#, collection_name=collection_name
    )
    DB.filterCollections(contains='first_run-main')#'-main-'
    for coll_name in DB.coll_names:
        date_str_initial = '-'.join(coll_name.split('-')[-2:])
        date = Utils.dateTime(date_str_initial)
        date_str = Utils.fileString(date)
        print(coll_name)
        print(date)
        print(date_str)
        print()
        if date_str_initial==date_str:
            continue
        DB.changeCollectionTitles([coll_name], date_str=date_str)
```
